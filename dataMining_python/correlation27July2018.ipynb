# Correlation

```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv("~/Desktop/marketValuePlayer/epldata_final.csv")
```

```python
data=pd.DataFrame(data1)
```

## dataset from kaggle of English Premier League Player Dataset

```python
data
```

```python
col1=data.loc[:,"fpl_points"]
```

```python
col2=data.loc[:,"market_value"]
```

```python
col1Average=col1.mean()
col2Average=col2.mean()
col1StdDev=col1.std()
col2StdDev=col2.std()
```

```python
col1Average
```

```python
correlationCoefficient=(((col1-col1Average)*(col2-col2Average)).sum())/(len(col1.index-1)*col1StdDev*col2StdDev)
```

```python
correlationCoefficient
```

## conclusion
        0<correlationCoefficient<1
        positively correlated

##  method 2:correlation chi-squared 

```python
column_names=["red","yellow","Green","Blue"]
row_names=["Introvert","Extrovert"]
```

```python
matrix= np.reshape((20 ,6 ,30 ,44 ,180 ,34 ,50 ,36 ), ( 2,4))
```

```python
matrix
```

```python
data2=pd.DataFrame(matrix,columns=column_names,index=row_names)
```

```python
data2
```

```python
data2['Total']=data2.sum(axis=1)
```

```python
data2
```

```python
data2.loc['Total']=data2.sum()
```

```python
data2
```

```python
n=data2.loc["Total","Total"]
n
```

```python
dataExpected=data2.copy()
dataExpected.iloc[1,1]=0
data2
```

```python
type(data2)
```

```python
dataExpected
```

```python
for i in range(len(data2.index)):
    for j in range(len(data2.columns)):
        dataExpected.iloc[i,j]=data2.iloc[-1,j]*data2.iloc[i,-1]/n
```

```python
dataExpected
```

```python
data2
```

```python
chiSquared=(data2-dataExpected)/dataExpected
```

```python
chiSquared
```

```python
chiSquared.iloc[:2,:4].values.sum()
```

## conclusion
        chiSquared>0
            positively correlated
